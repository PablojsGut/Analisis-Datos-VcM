```python
df = df_init.copy()
```

```python
# Lista de columnas a eliminar
columnas_a_eliminar = [
    'Hora de inicio',
    'Hora de finalización',
    'Correo electrónico',
    'Nombre Correo',
    'Nombre',
    'Género',
    'Sexo',
    'Actividad Alumni',
    'Años Actividad Alumni',
    'Descripción Actividad Alumni',
    'Cant. participantes',
    'Nombre Actor externo Alumni',
    'Nombre Mesa',
    'Nombre Actor externo Mesa',
    'Descripción Mesa',
    'Nombre del Congreso, Seminario o Coloquio',
    'Entidad Organizadora',
    'N° Estudiantes participantes (HOMBRES)',
    'N° Estudiantes participantes (MUJERES)',
    'N° docentes participantes (HOMBRES)',
    'N° docentes participantes (MUJERES)',
    'Nombre del programa de corta duración',
    'Nombre de la institución anfitriona',
    'Objetivo de la pasantía/estancia',
    'Nombre del académico/a visitante',
    'Institución de origen',
    'Nombre Actividad',
    'Año Actividad',
    'Ciudad Actividad',
    'País Actividad',
    'Objetivo Actividad',
    'Asignatura asociada',
    'Docente responsable Actividad',
    'Nombre Actor externo Actividad',
    'N° Estudiantes participantes en la salida (HOMBRES)',
    'N° Estudiantes participantes en la salida (MUJERES)',
    'Retroalimentación',
    'Documentos'
]

# Eliminar las columnas directamente del DataFrame
# Usamos errors='ignore' para evitar que falle si alguna columna ya fue eliminada o renombrada
df.drop(columns=columnas_a_eliminar, inplace=True, errors='ignore')

print("--- Columnas eliminadas exitosamente. ---")
print(f"\nNúmero de columnas restantes: {len(df.columns)}")
print("\nLista de columnas restantes:")
print(df.columns.tolist())
```

```python
# 1. Definir la lista de columnas que se consolidarán en 'Subdependencia'
columnas_subdependencia = [
    'Unidad No Académica',
    'Programa de Magíster',
    'Núcleos de Investigación',
    'Centros de Investigación',
    'Programa de Doctorado',
    'Escuela/Carrera Facultad de Ciencias Sociales y Artes',
    'Escuela/Carrera Facultad de Ciencias, Ingeniería y Tecnología',
    'Escuela/Carrera Medicina y Ciencias de la Salud',
    'Especialidad Odontológica',
    'Especialidad Médica'
]

# 2. Crear la columna 'Subdependencia'
# El método .bfill(axis=1) rellena los nulos de izquierda a derecha usando el siguiente valor válido.
# Esto asegura que el único valor no nulo de la fila "suba" a la nueva columna.
df['Subdependencia'] = df[columnas_subdependencia].bfill(axis=1).iloc[:, 0]

# Opcional: Reemplazar NaN en 'Subdependencia' con una categoría de control
# Esto es útil si quieres saber cuáles registros no tenían Subdependencia específica.
df['Subdependencia'].fillna('No Aplica', inplace=True)

# 3. Insertar la columna 'Subdependencia' después de 'Dependencia'
# Identificamos el índice de 'Dependencia'
idx_dependencia = df.columns.get_loc('Dependencia')

# Obtenemos la lista actual de columnas
cols = df.columns.tolist()

# Removemos 'Subdependencia' de su posición final y la insertamos después de 'Dependencia'
cols.remove('Subdependencia')
cols.insert(idx_dependencia + 1, 'Subdependencia')

# Aplicamos el nuevo orden al DataFrame
df = df[cols]

print("--- Columna 'Subdependencia' creada exitosamente. ---")
print(df[['Dependencia', 'Subdependencia']].head().to_markdown())
```

```python
# Eliminar las columnas originales después de la consolidación
df.drop(columns=columnas_subdependencia, inplace=True, errors='ignore')

print("\n--- Columnas originales de subdependencia eliminadas. ---")
print(f"Número de columnas restantes: {len(df.columns)}")
print("\nLista de columnas finales:")
print(df.columns.tolist())
```

```python
# 1. Asegurar que las columnas de fecha sean de tipo datetime
# Usamos errors='coerce' para convertir cualquier valor no válido (texto o nulo) a NaT (Not a Time).
df['Fecha inicio Actividad'] = pd.to_datetime(df['Fecha inicio Actividad'], errors='coerce')
df['Fecha termino Actividad'] = pd.to_datetime(df['Fecha termino Actividad'], errors='coerce')

# 2. Calcular la diferencia y extraer los días
# El resultado es la duración en días (valor entero)
df['Total Días Actividad'] = (
    (df['Fecha termino Actividad'] - df['Fecha inicio Actividad']) + pd.Timedelta(days=1)
).dt.days

# 3. Limpieza de la nueva columna
# Asumimos que si la diferencia es nula (porque faltaba alguna fecha), la actividad duró 0 días.
df['Total Días Actividad'].fillna(0, inplace=True)

# 4. Reordenar la columna para ubicarla después de 'Fecha termino Actividad'
# a. Encontrar el índice de 'Fecha termino Actividad'
idx_fecha_termino = df.columns.get_loc('Fecha termino Actividad')

# b. Obtener la lista actual de columnas
cols = df.columns.tolist()

# c. Mover 'Total Días' a la posición correcta si no está ya ahí
if cols[-1] == 'Total Días Actividad':
    cols.remove('Total Días Actividad')
    cols.insert(idx_fecha_termino + 1, 'Total Días Actividad')
    df = df[cols]
# Si ya estaba en la posición, no se hace nada.

print("--- Columna 'Total Días Actividad' creada y reordenada exitosamente. ---")
print(f"\nEstadísticas de la nueva columna:\n{df['Total Días Actividad'].describe().to_markdown()}")

print("\nPrimeras filas de las columnas de fecha y duración:")
print(df[['Fecha inicio Actividad', 'Fecha termino Actividad', 'Total Días Actividad']].head().to_markdown())

print("\nLista de columnas finales:")
print(df.columns.tolist())
```

```python
# 1. Convertir las columnas de fecha a formato datetime
df['Fecha inicio Actividad'] = pd.to_datetime(df['Fecha inicio Actividad'])
df['Fecha termino Actividad'] = pd.to_datetime(df['Fecha termino Actividad'])

# 2. Identificar filas donde la fecha de inicio es POSTERIOR a la de término
swap_mask = df['Fecha inicio Actividad'] > df['Fecha termino Actividad']

# 3. Realizar el intercambio (swap) de las fechas SÓLO en esas filas
# Esto corrige los datos de origen
df.loc[swap_mask, ['Fecha inicio Actividad', 'Fecha termino Actividad']] = \
    df.loc[swap_mask, ['Fecha termino Actividad', 'Fecha inicio Actividad']].values

# 4. Recalcular 'Total Días Actividad' con las fechas corregidas
# Usamos .dt.days + 1 para incluir el día de inicio y el de término (como sugiere tu head data)
df['Total Días Actividad'] = (df['Fecha termino Actividad'] - df['Fecha inicio Actividad']).dt.days + 1
print(df.head().to_markdown())
```

```python
target_col = 'Total Días Actividad'
nan_count = df[target_col].isnull().sum()

print(f"Total de valores NaN en '{target_col}': {nan_count}")

if nan_count > 0:
    print(f"\nFilas con NaN en '{target_col}' (primeras 10):")
    # Muestra las columnas que causan el problema: las fechas originales y el resultado
    cols_to_show = ['Fecha inicio Actividad', 'Fecha termino Actividad', target_col]
    
    # Filtramos y mostramos las filas donde 'Total Días Actividad' es nulo
    nan_rows = df[df[target_col].isnull()]
    print(nan_rows[cols_to_show].head(10).to_markdown())
else:
    print("\nNo se encontraron valores NaN en la columna de destino.")
```

```python
# --- 1. Limpieza de NaN y separación de valores ---

# 1.2 Separar la cadena de texto en listas de ámbitos. 
# Usamos .str.split(';') y luego .explode() para que cada ámbito esté en una fila distinta.
# Luego usamos .str.strip() para eliminar espacios en blanco residuales.
df_ambitos_exploded = df['Ámbitos Estratégicos Actividad'].str.split(';').explode().str.strip()

# 1.3 Filtrar valores vacíos que puedan resultar de ';' al final o dobles delimitadores
df_ambitos_exploded = df_ambitos_exploded[df_ambitos_exploded != '']

# --- 2. Aplicar One-Hot Encoding ---

# 2.1 Usar pd.get_dummies para crear las columnas binarias
# Establecemos prefix='Ámbitos Estratégicos' para el nombre de las nuevas columnas
df_dummies = pd.get_dummies(
    df_ambitos_exploded, 
    prefix='Ámbitos Estratégicos'
)

# 2.2 Agrupar por el índice original para sumar las codificaciones (True/False)
# La suma de las columnas True/False por el índice original nos da el resultado final.
# Si la fila tenía un ámbito, la suma será 1 (True), si no, 0 (False).
df_dummies_grouped = df_dummies.groupby(df_dummies.index).sum()

# 2.3 Convertir de 1/0 a True/False (opcional, pero más claro para 'dummies')
# Convertimos los 1s a True y 0s a False
df_dummies_bool = df_dummies_grouped.astype(bool)

# --- 3. Combinar con el DataFrame original ---

# Unir las nuevas columnas al DataFrame de regresión lineal
df = df.merge(
    df_dummies_bool, 
    left_index=True, 
    right_index=True, 
    how='left'
)

df.drop(columns=['Ámbitos Estratégicos Actividad'], inplace=True)

print("--- Primeras 5 filas con las nuevas columnas de Ámbitos ---")

# Mostrar las columnas originales de fecha, la duración, y las nuevas columnas binarias
cols_to_display = ['Fecha inicio Actividad', 'Fecha termino Actividad', 'Total Días Actividad'] + list(df_dummies_bool.columns)[:5]
print(df[cols_to_display].head().to_markdown())
print(f"DataFrame con {len(df_dummies_bool.columns)} nuevas columnas binarias.\n")
print(df.head().to_markdown())
```

```python
# 1. Identificar todas las columnas que representan los Ámbitos Estratégicos
ambito_cols = [col for col in df.columns if col.startswith('Ámbitos Estratégicos_')]

# 2. Contar la cantidad de TRUEs (o 1s) en esas columnas para cada fila
# .sum(axis=1) suma horizontalmente a través de las columnas identificadas.
# El resultado será la nueva columna 'Total Ámbitos Estratégicos'.
df['Total Ámbitos Estratégicos'] = df[ambito_cols].sum(axis=1)

print("Columna 'Total Ámbitos Estratégicos' creada exitosamente.")
print(f"Rango de conteo: {df['Total Ámbitos Estratégicos'].min()} a {df['Total Ámbitos Estratégicos'].max()}")

# 3. Mostrar la nueva columna junto con la duración
cols_to_display = ['Total Días Actividad', 'Total Ámbitos Estratégicos'] + list(ambito_cols)[:3]
print("\n--- Vista de la nueva columna de conteo ---")
print(df[cols_to_display].head().to_markdown())
```

```python
# Usamos .str.split(';') y luego .explode() para que cada ODS esté en una fila distinta.
df_ods_exploded = df['Objetivos de Desarrollo Sostenible (ODS) Actividad'].str.split(';').explode().str.strip()

# 1.3 Filtrar valores vacíos
df_ods_exploded = df_ods_exploded[df_ods_exploded != '']

# --- 2. Limpieza del Prefijo Numérico ---

# Usaremos una expresión regular para limpiar el prefijo 'NÚMERO: ' de cada ODS
# Ejemplo: '13: Acción por el Clima' -> 'Acción por el Clima'
# También capturará y mantendrá 'No Aplica' o cualquier otro valor sin el patrón 'Número: '
df_ods_cleaned = df_ods_exploded.str.replace(r'^\d{1,2}:\s*', '', regex=True).str.strip()

# --- 3. Aplicar One-Hot Encoding ---

# 3.1 Usar pd.get_dummies para crear las columnas binarias
df_ods_dummies = pd.get_dummies(
    df_ods_cleaned, 
    prefix='ODS'
)

# 3.2 Agrupar por el índice original para sumar las codificaciones (1/0)
df_ods_grouped = df_ods_dummies.groupby(df_ods_dummies.index).sum()

# 3.3 Convertir de 1/0 a True/False
df_ods_bool = df_ods_grouped.astype(bool)

# --- 4. Combinar con el DataFrame original y Limpieza ---

# 4.1 Unir las nuevas columnas al DataFrame principal
df = df.merge(
    df_ods_bool, 
    left_index=True, 
    right_index=True, 
    how='left'
)

# 4.3 Eliminar la columna original
df.drop(columns=['Objetivos de Desarrollo Sostenible (ODS) Actividad'], inplace=True)

print(f"DataFrame con {len(df_ods_bool.columns)} nuevas columnas binarias de ODS.\n")
print("--- Primeras 5 filas con las nuevas columnas de ODS ---")

# Mostrar las columnas de duración y las 5 primeras nuevas columnas binarias de ODS
cols_ods_display = ['Total Días Actividad'] + list(df_ods_bool.columns)[:5]
print(df[cols_ods_display].head().to_markdown())

print("\n--- Vista completa de las primeras 5 filas del DataFrame final (df) ---")
print(df.head().to_markdown())
```

```python
# 1. Identificar todas las columnas que representan los ODS.
# Las creamos en el paso anterior y comienzan con 'ODS_'.
ods_cols = [col for col in df.columns if col.startswith('ODS_')]

# 2. Contar la cantidad de TRUEs (o 1s) en esas columnas para cada fila
# .sum(axis=1) suma las ocurrencias de ODS para cada registro.
# El resultado será la nueva columna 'Total ODS'.
df['Total ODS'] = df[ods_cols].sum(axis=1)

print("Columna 'Total ODS' creada exitosamente.")
print(f"Rango de conteo: {df['Total ODS'].min()} a {df['Total ODS'].max()}")

# 3. Mostrar la nueva columna junto con la duración y las 3 primeras columnas binarias
cols_to_display = ['Total Días Actividad', 'Total ODS'] + list(ods_cols)[:3]
print("\n--- Vista de la nueva columna de conteo ---")
print(df[cols_to_display].head().to_markdown())
```

```python

df_sede_exploded = df['Sede'].str.split(';').explode().str.strip()

# 1.3 Filtrar valores vacíos
df_sede_exploded = df_sede_exploded[df_sede_exploded != '']


# --- 3. Aplicar One-Hot Encoding ---

# 3.1 Usar pd.get_dummies para crear las columnas binarias
df_sede_dummies = pd.get_dummies(
    df_sede_exploded, 
    prefix='Sede',
    dtype=bool 
)

# 3.2 Agrupar por el índice original para sumar las codificaciones (1/0)
df_sede_grouped = df_sede_dummies.groupby(df_sede_dummies.index).sum()

# 3.3 Convertir de 1/0 a True/False
df_sede_bool = df_sede_grouped.astype(bool)

# --- 4. Combinar con el DataFrame original y Limpieza ---

# 4.1 Unir las nuevas columnas al DataFrame principal
df = df.merge(
    df_sede_bool, 
    left_index=True, 
    right_index=True, 
    how='left'
)

# 4.3 Eliminar la columna original
df.drop(columns=['Sede'], inplace=True)

print(f"DataFrame con {len(df_sede_bool.columns)} nuevas columnas binarias de Sede.\n")

# Mostrar las columnas de duración y las 5 primeras nuevas columnas binarias de Sede
sede_cols_display = ['Total Días Actividad'] + list(df_sede_bool.columns)[:5]
print("--- Primeras 5 filas con las nuevas columnas de Sede ---")
print(df[sede_cols_display].head().to_markdown())
```
